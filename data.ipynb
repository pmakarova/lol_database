```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

def setup_driver():
    """Настраивает и возвращает драйвер Chrome"""
    options = Options()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--window-size=1920,1080')
    driver = webdriver.Chrome(options=options)
    return driver

def scroll_page(driver):
    """Прокручивает страницу для загрузки всего контента"""
    for i in range(5):
        driver.execute_script(f"window.scrollTo(0, {500 * (i + 1)});")
        time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)


def get_champion_lore(champ, driver=None):
    """
    Получает лор чемпиона
    
    Args:
        champion_name (str): Имя чемпиона
        driver: Существующий драйвер (опционально)
    
    Returns:
        stories: Словарь с данными лора или None если лор не найден
    """
    story_url = "https://universe.leagueoflegends.com/en_US/story/champion/{}/"
    
    own_driver = driver is None
    if own_driver:
        driver = setup_driver()
        
    try:
        url = story_url.format(champ.lower())
        print(f"Пытаемся получить: {url}")
        
        driver.get(url)
        time.sleep(3)
        scroll_page(driver)
        
        soup = BeautifulSoup(driver.page_source, "html.parser")
        
        # Заголовок находится в noHeaderSubtitle_1Rct
        title_element = soup.find("div", class_="noHeaderSubtitle_1Rct")
        title = title_element.get_text(strip=True) if title_element else f"Story of {champ}"
        
        # Поиск основного текста
        story_text = ""
        all_paragraphs = soup.find_all("p")
        content_paragraphs = []
        
        for p in all_paragraphs:
            text = p.get_text(strip=True)
            # Фильтруем ненужные строки
            if (text and len(text) > 20 and
                "©" not in text and 
                "Riot Games" not in text and
                "League of Legends" not in text and
                "Scroll to Begin" not in text and
                "Champions" not in text and
                "Параметры хранения данных" not in text):  # <- убираем конкретную строку
                content_paragraphs.append(text)
        
        if content_paragraphs:
            story_text = "\n".join(content_paragraphs)
        
        # Сохраняем результат
        if story_text and len(story_text) > 100:
            story = ({
                'champion': champ,
                'title': title,
                'text': story_text[:200],
                'length': len(story_text)
            })
            print(f"\n✓ {champ.upper()} - '{title}'")
            print(f"  Длина: {len(story_text)} символов")
            print(f"  Превью: {story_text[:200]}...")

            return story
        else:
            print(f"❌ Не найден лор для {champ}")

    except Exception as e:
        print(f"❌ Ошибка при получении лора для {champ}: {e}")
        return None
    
    finally:
        if own_driver:
            driver.quit()
```

```python
from datetime import datetime
import re
def parse_date(date_str):
    """
    Преобразует дату из формата в 'YYYY-MM-DD'
    """
    date_str = re.sub(r'(\d+)(st|nd|rd|th)', r'\1', date_str)
    
    date_obj = datetime.strptime(date_str, '%d %B %Y')
    
    return date_obj.strftime('%Y-%m-%d')
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def get_team_rosters_link(base_url):
    response = requests.get(base_url)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')

    top_tabs = soup.select("div.tabheader-top div.tabheader-tab a")

    if len(top_tabs) < 2:
        print("Недостаточно вкладок верхнего уровня")
        return None
    
    second_tab_href = top_tabs[1]["href"]
    second_tab_url = urljoin(base_url, second_tab_href)

    response2 = requests.get(second_tab_url)
    response2.raise_for_status()
    soup2 = BeautifulSoup(response2.text, "html.parser")

    sub_tabs = soup2.select("div.tabheader-top div.tabheader-tab a")

    if not sub_tabs:
        print("❌ Вкладки не найдены.")
        return None

    team_roster_link = None
    for a in sub_tabs:
        if "Team_Rosters" in a["href"]:
            team_roster_link = urljoin(second_tab_url, a["href"])
            break
        
    return team_roster_link
```

```python
def clean_all_datasets():
    """Очищает все наборы данных"""
    
    # Summoner
    cleaned_summoner = []
    for row in summoner_values:
        cleaned_row = list(row)
        cleaned_row[1] = clean_text_advanced(row[1])  # имя summoner
        cleaned_summoner.append(tuple(cleaned_row))
    
    # Lore
    cleaned_lore = []
    for row in lore_values:
        cleaned_row = tuple(clean_text_advanced(item) if isinstance(item, str) else item for item in row)
        cleaned_lore.append(cleaned_row)
    
    # Champion
    cleaned_champ = []
    for row in champ_values:
        cleaned_row = list(row)
        cleaned_row[2] = clean_text_advanced(row[2])  # описание чемпиона
        cleaned_champ.append(tuple(cleaned_row))
    
    # Skin - добавляем обработку pandas.NA и очистку текста
    cleaned_skin = []
    for row in skins_values:
        cleaned_row = []
        for value in row:
            if pd.isna(value):
                cleaned_row.append(None)
            elif isinstance(value, str):
                cleaned_row.append(clean_text_advanced(value))
            else:
                cleaned_row.append(value)
        cleaned_skin.append(tuple(cleaned_row))
    
    return cleaned_summoner, cleaned_lore, cleaned_champ, cleaned_skin

# Использование
cleaned_summoner_values, cleaned_lore_values, cleaned_champ_values, cleaned_skin_values = clean_all_datasets()
```
